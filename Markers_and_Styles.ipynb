```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
```

Example 1: Basic Line with Markers

```python
# Plot with a basic line and markers at each data point
plt.plot(x, y, marker='o', linestyle='-', color='b')

# Title and labels
plt.title('Basic Line with Markers')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

Example 2: Customizing Line Styles and Markers

```python

# Plot with different line styles, markers, and colors
plt.plot(x, y, marker='s', linestyle='--', color='r', linewidth=2, markersize=8)

# Title and labels
plt.title('Custom Line Style and Markers')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

Example 3: Multiple Lines with Different Styles and Colors

```python
# Data for additional lines
y2 = [1, 4, 6, 8, 10]
y3 = [2, 4, 4, 6, 8]

# Plotting multiple lines with different styles
plt.plot(x, y, marker='o', linestyle='-', color='g', label='Line 1')
plt.plot(x, y2, marker='^', linestyle='--', color='m', label='Line 2')
plt.plot(x, y3, marker='s', linestyle='-.', color='c', label='Line 3')

# Adding title, labels, and legend
plt.title('Multiple Lines with Custom Styles')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Show the plot
plt.show()
```

Example 4: Customizing Line Widths and Transparency

```python
# Plot with custom line width and transparency
plt.plot(x, y, marker='D', linestyle='-', color='b', linewidth=3, alpha=0.7)

# Title and labels
plt.title('Line Width and Transparency')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

Example 5: Changing Marker Colors and Line Styles Separately

```python
# You can separate the marker color from the line color
plt.plot(x, y, marker='o', linestyle='-', color='orange', markerfacecolor='blue', markeredgecolor='black', markersize=10)

# Title and labels
plt.title('Separate Marker and Line Colors')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

Example 1: Basic Line Plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the line plot
x = np.linspace(0, 10, 100)  # 100 points between 0 and 10
y = 2 * x + 1  # Linear relationship

# Create a basic line plot
plt.plot(x, y)

# Add title and labels
plt.title('Basic Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

Example 2: Customizing Line Styles and Colors

```python
# Create a line plot with customized line style, color, and marker
plt.plot(x, y, linestyle='--', color='r', marker='o', linewidth=2, markersize=5)

# Add title and labels
plt.title('Customized Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

Example 3: Adding Grid, Legends, and Customizing Line Width

```python
# Data for another line plot
y2 = 0.5 * x ** 2  # Quadratic relationship

# Plot multiple lines
plt.plot(x, y, linestyle='-', color='b', marker='s', label='Linear')
plt.plot(x, y2, linestyle='--', color='g', marker='^', label='Quadratic')

# Add grid, title, labels, and legend
plt.grid(True)
plt.title('Multiple Lines with Customizations')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Show the plot
plt.show()
```

 Example 4: Using Transparency and Custom Marker Styles

```python
# Create a line plot with custom marker styles and transparency
plt.plot(x, y, linestyle=':', color='m', marker='D', markerfacecolor='yellow', markeredgecolor='black', markersize=8, alpha=0.7)

# Add title and labels
plt.title('Line Plot with Custom Markers and Transparency')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```
