```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.widgets import Button

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
y_cos = np.cos(x)
```

1. Annotation Example

```python
plt.figure(figsize=(8, 6))

# Plot the data
plt.plot(x, y, label='Sine Wave')

# Annotate a specific point
plt.annotate('Local Max', 
             xy=(np.pi/2, np.sin(np.pi/2)), 
             xytext=(np.pi/2 + 1, np.sin(np.pi/2) + 0.5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=12,
             color='red')

# Add title and labels
plt.title('Sine Wave with Annotation')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

2. Text Example

```python
plt.figure(figsize=(8, 6))

# Plot the data
plt.plot(x, y_cos, label='Cosine Wave')

# Add text at a specific location
plt.text(5, 0, 'Cosine Curve', fontsize=14, color='blue', ha='center')

# Add title and labels
plt.title('Cosine Wave with Text')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

3. Interactive Button Example

```python
fig, ax = plt.subplots(figsize=(8, 6))
plt.subplots_adjust(bottom=0.2)  # Adjust space for the button

# Plot the data
line, = plt.plot(x, y, label='Sine Wave')

# Button callback function
def toggle_line(event):
    if line.get_visible():
        line.set_visible(False)
    else:
        line.set_visible(True)
    plt.draw()

# Add a button for interaction
ax_button = plt.axes([0.4, 0.05, 0.2, 0.075])  # Position of the button
button = Button(ax_button, 'Toggle Line')
button.on_clicked(toggle_line)

# Add title and labels
plt.title('Sine Wave with Interactive Button')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

### Working with color maps and palettes.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
X, Y = np.meshgrid(x, x)
Z = np.sin(X) * np.cos(Y)
```

Using Built-in Color Maps

```python
plt.figure(figsize=(8, 6))

# Plot with a built-in color map
plt.imshow(Z, cmap='viridis', interpolation='nearest')
plt.colorbar(label='Intensity')  # Show color scale

# Add title and labels
plt.title('Color Map with Viridis')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

2. Creating Custom Color Maps

```python
from matplotlib.colors import LinearSegmentedColormap

# Define a custom color map
colors = ['blue', 'green', 'yellow', 'red']
n_bins = [3]  # Discretizes the interpolation into bins
cmap_name = 'custom_map'
custom_cmap = LinearSegmentedColormap.from_list(cmap_name, colors, N=256)

plt.figure(figsize=(8, 6))

# Plot with the custom color map
plt.imshow(Z, cmap=custom_cmap, interpolation='nearest')
plt.colorbar(label='Intensity')  # Show color scale

# Add title and labels
plt.title('Custom Color Map')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

3. Applying Color Maps to Different Plot Types
    Scatter plot with color map

```python
plt.figure(figsize=(8, 6))

# Data for scatter plot
scatter_x = np.random.rand(100)
scatter_y = np.random.rand(100)
colors = np.sqrt(scatter_x**2 + scatter_y**2)  # Color by distance from origin

plt.scatter(scatter_x, scatter_y, c=colors, cmap='plasma', edgecolor='k')
plt.colorbar(label='Distance from Origin')  # Show color scale

# Add title and labels
plt.title('Scatter Plot with Plasma Color Map')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.grid(True)
plt.show()
```
