## Creating multiple subplots using plt.subplot() and plt.subplots().

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x / 10)
```

Example 1: Creating Subplots with plt.subplot()

```python
# Create a figure
plt.figure(figsize=(10, 8))

# First subplot (1 row, 2 columns, 1st subplot)
plt.subplot(2, 2, 1)  # (rows, columns, position)
plt.plot(x, y1, 'r')
plt.title('Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

# Second subplot (1 row, 2 columns, 2nd subplot)
plt.subplot(2, 2, 2)
plt.plot(x, y2, 'g')
plt.title('Cosine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

# Third subplot (1 row, 2 columns, 3rd subplot)
plt.subplot(2, 2, 3)
plt.plot(x, y3, 'b')
plt.title('Tangent Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

# Fourth subplot (1 row, 2 columns, 4th subplot)
plt.subplot(2, 2, 4)
plt.plot(x, y4, 'm')
plt.title('Exponential Growth')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

Example 2: Creating Subplots with plt.subplots()

```python
# Create a figure and axes array
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# First subplot
axs[0, 0].plot(x, y1, 'r')
axs[0, 0].set_title('Sine Wave')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')
axs[0, 0].grid(True)

# Second subplot
axs[0, 1].plot(x, y2, 'g')
axs[0, 1].set_title('Cosine Wave')
axs[0, 1].set_xlabel('X-axis')
axs[0, 1].set_ylabel('Y-axis')
axs[0, 1].grid(True)

# Third subplot
axs[1, 0].plot(x, y3, 'b')
axs[1, 0].set_title('Tangent Wave')
axs[1, 0].set_xlabel('X-axis')
axs[1, 0].set_ylabel('Y-axis')
axs[1, 0].grid(True)

# Fourth subplot
axs[1, 1].plot(x, y4, 'm')
axs[1, 1].set_title('Exponential Growth')
axs[1, 1].set_xlabel('X-axis')
axs[1, 1].set_ylabel('Y-axis')
axs[1, 1].grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

## Adjusting subplot layouts with plt.tight_layout().

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x / 10)
```

Example 3: Using plt.tight_layout() with Padding Adjustment

```python
# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(8, 6))

# Plot on each subplot
axs[0, 0].plot(x, y1, 'r')
axs[0, 0].set_title('Sine Wave')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')

axs[0, 1].plot(x, y2, 'g')
axs[0, 1].set_title('Cosine Wave')
axs[0, 1].set_xlabel('X-axis')
axs[0, 1].set_ylabel('Y-axis')

axs[1, 0].plot(x, y3, 'b')
axs[1, 0].set_title('Tangent Wave')
axs[1, 0].set_xlabel('X-axis')
axs[1, 0].set_ylabel('Y-axis')

axs[1, 1].plot(x, y4, 'm')
axs[1, 1].set_title('Exponential Growth')
axs[1, 1].set_xlabel('X-axis')
axs[1, 1].set_ylabel('Y-axis')

# Adjust layout with custom padding between plots
plt.tight_layout(pad=3.0)  # Increasing padding between subplots

# Show the plot with customized layout
plt.show()
```

## Customizing each subplot independently.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x / 3)
```

Example 1: Customizing Each Subplot with plt.subplot()

```python
# Create a figure
plt.figure(figsize=(10, 8))

# First subplot (customized independently)
plt.subplot(2, 2, 1)
plt.plot(x, y1, marker='o', linestyle='-', color='r', label='Sine')
plt.title('Sine Wave', fontsize=12, color='r')
plt.xlabel('X-axis', fontsize=10)
plt.ylabel('Y-axis', fontsize=10)
plt.grid(True)
plt.legend()

# Second subplot (customized independently)
plt.subplot(2, 2, 2)
plt.plot(x, y2, marker='s', linestyle='--', color='g', label='Cosine')
plt.title('Cosine Wave', fontsize=12, color='g')
plt.xlabel('X-axis', fontsize=10)
plt.ylabel('Y-axis', fontsize=10)
plt.grid(True)
plt.legend()

# Third subplot (customized independently)
plt.subplot(2, 2, 3)
plt.plot(x, y3, marker='D', linestyle=':', color='b', label='Tangent')
plt.title('Tangent Wave', fontsize=12, color='b')
plt.xlabel('X-axis', fontsize=10)
plt.ylabel('Y-axis', fontsize=10)
plt.grid(True)
plt.legend()

# Fourth subplot (customized independently)
plt.subplot(2, 2, 4)
plt.plot(x, y4, marker='^', linestyle='-', color='m', label='Exponential')
plt.title('Exponential Growth', fontsize=12, color='m')
plt.xlabel('X-axis', fontsize=10)
plt.ylabel('Y-axis', fontsize=10)
plt.grid(True)
plt.legend()

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

Example 2: Customizing Each Subplot with plt.subplots()

```python

# Create figure and subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# First subplot
axs[0, 0].plot(x, y1, marker='o', linestyle='-', color='r')
axs[0, 0].set_title('Sine Wave', fontsize=12, color='r')
axs[0, 0].set_xlabel('X-axis', fontsize=10)
axs[0, 0].set_ylabel('Y-axis', fontsize=10)
axs[0, 0].grid(True)

# Second subplot
axs[0, 1].plot(x, y2, marker='s', linestyle='--', color='g')
axs[0, 1].set_title('Cosine Wave', fontsize=12, color='g')
axs[0, 1].set_xlabel('X-axis', fontsize=10)
axs[0, 1].set_ylabel('Y-axis', fontsize=10)
axs[0, 1].grid(True)

# Third subplot
axs[1, 0].plot(x, y3, marker='D', linestyle=':', color='b')
axs[1, 0].set_title('Tangent Wave', fontsize=12, color='b')
axs[1, 0].set_xlabel('X-axis', fontsize=10)
axs[1, 0].set_ylabel('Y-axis', fontsize=10)
axs[1, 0].grid(True)

# Fourth subplot
axs[1, 1].plot(x, y4, marker='^', linestyle='-', color='m')
axs[1, 1].set_title('Exponential Growth', fontsize=12, color='m')
axs[1, 1].set_xlabel('X-axis', fontsize=10)
axs[1, 1].set_ylabel('Y-axis', fontsize=10)
axs[1, 1].grid(True)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

## Types of Plots 

```python


import matplotlib.pyplot as plt
import numpy as np

# Sample data for different types of plots
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.random.randn(100)
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 2, 5]

#------------------------------------------------------------
# Create a figure with 4 subplots (2 rows, 2 columns)
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# First subplot: Line plot
axs[0, 0].plot(x, y1, marker='o', linestyle='-', color='r')
axs[0, 0].set_title('Line Plot: Sine Wave', fontsize=12, color='r')
axs[0, 0].set_xlabel('X-axis', fontsize=10)
axs[0, 0].set_ylabel('Y-axis', fontsize=10)
axs[0, 0].grid(True)

#------------------------------------------------------------
# Second subplot: Scatter plot
axs[0, 1].scatter(x, y2, marker='x', color='g')
axs[0, 1].set_title('Scatter Plot: Cosine Wave', fontsize=12, color='g')
axs[0, 1].set_xlabel('X-axis', fontsize=10)
axs[0, 1].set_ylabel('Y-axis', fontsize=10)
axs[0, 1].grid(True)

#------------------------------------------------------------
# Third subplot: Histogram
axs[1, 0].hist(y3, bins=20, color='b', edgecolor='black')
axs[1, 0].set_title('Histogram: Random Data', fontsize=12, color='b')
axs[1, 0].set_xlabel('Bins', fontsize=10)
axs[1, 0].set_ylabel('Frequency', fontsize=10)
axs[1, 0].grid(True)

#------------------------------------------------------------
# Fourth subplot: Bar plot
axs[1, 1].bar(categories, values, color='m', edgecolor='black')
axs[1, 1].set_title('Bar Plot: Category Values', fontsize=12, color='m')
axs[1, 1].set_xlabel('Categories', fontsize=10)
axs[1, 1].set_ylabel('Values', fontsize=10)
axs[1, 1].grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the figure
plt.show()
```
