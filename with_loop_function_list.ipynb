### Functions

```python
import matplotlib.pyplot as plt

# Define a function to create a plot
def plot_line(x, y, title):
    plt.plot(x, y, marker='o', linestyle='-', color='b')
    plt.title(title)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)
    plt.show()

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Call the function with data
plot_line(x, y, 'Simple Line Plot')
```

### Loops

```python
import matplotlib.pyplot as plt

# Data for multiple lines
x = [1, 2, 3, 4, 5]
ys = [[2, 4, 6, 8, 10], [1, 3, 5, 7, 9], [0, 2, 4, 6, 8]]

# Plot each line
for y in ys:
    plt.plot(x, y, marker='o')

plt.title('Multiple Lines Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.legend(['Line 1', 'Line 2', 'Line 3'])
plt.show()
```

### Lists

```python
import matplotlib.pyplot as plt

# Data in lists
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a scatter plot
plt.scatter(x, y, color='r', marker='x')
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```
