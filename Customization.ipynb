### Adding titles, labels, and legends to plots.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

Example 1: Adding Titles to a Plot

```python
# Create a plot with a title
plt.plot(x, y1, color='b')
plt.title('Sine Wave')  # Adding a title

# Show the plot
plt.show()
```

Example 2: Adding Axis Labels

```python
# Create a plot with axis labels
plt.plot(x, y2, color='r')
plt.title('Cosine Wave')  # Adding a title
plt.xlabel('X-axis')      # Label for X-axis
plt.ylabel('Y-axis')      # Label for Y-axis

# Show the plot
plt.show()
```

Example 3: Adding a Legend

```python
# Plot multiple lines with a legend
plt.plot(x, y1, color='b', linestyle='-', label='Sine')  # Line 1
plt.plot(x, y2, color='r', linestyle='--', label='Cosine')  # Line 2

# Add title, labels, and legend
plt.title('Sine and Cosine Waves')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()  # Adding a legend to the plot

# Show the plot
plt.show()
```

Example 4: Customizing Legend Location

```python
# Plot multiple lines with customized legend location
plt.plot(x, y1, color='b', linestyle='-', label='Sine')
plt.plot(x, y2, color='r', linestyle='--', label='Cosine')

# Add title, labels, and a legend at a specific location
plt.title('Sine and Cosine Waves with Custom Legend Location')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc='upper right')  # Customizing legend location

# Show the plot
plt.show()
```

Example 5: Customizing Legend Appearance

```python
# Plot multiple lines with customized legend appearance
plt.plot(x, y1, color='b', linestyle='-', label='Sine')
plt.plot(x, y2, color='r', linestyle='--', label='Cosine')

# Add title, labels, and customize legend appearance
plt.title('Sine and Cosine Waves with Customized Legend')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc='best', fontsize='large', shadow=True)  # Customize legend font size and shadow

# Show the plot
plt.show()
```

### Customizing font sizes, colors, and styles.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

#------------------------------------------------------------
# Example 1: Customizing Font Sizes
#------------------------------------------------------------

# Create a plot with customized font sizes
plt.plot(x, y, color='b')
plt.title('Plot with Custom Font Sizes', fontsize=16)  # Title font size
plt.xlabel('X-axis', fontsize=14)  # X-axis label font size
plt.ylabel('Y-axis', fontsize=14)  # Y-axis label font size

# Show the plot
plt.show()
```

Example 2: Customizing Font Colors

```python
# Create a plot with customized font colors
plt.plot(x, y, color='r')
plt.title('Plot with Custom Font Colors', fontsize=16, color='purple')  # Title font color
plt.xlabel('X-axis', fontsize=14, color='green')  # X-axis label font color
plt.ylabel('Y-axis', fontsize=14, color='blue')  # Y-axis label font color

# Show the plot
plt.show()
```

Example 3: Customizing Font Styles

```python
# Create a plot with customized font styles
plt.plot(x, y, color='g')
plt.title('Plot with Custom Font Styles', fontsize=16, fontstyle='italic')  # Title font style
plt.xlabel('X-axis', fontsize=14, fontweight='bold')  # X-axis label font weight
plt.ylabel('Y-axis', fontsize=14, fontfamily='serif')  # Y-axis label font family

# Show the plot
plt.show()
```

Example 4: Combining Font Customizations

```python
# Create a plot with combined font customizations
plt.plot(x, y, color='m')
plt.title('Plot with Combined Font Customizations', fontsize=18, color='darkblue', fontweight='bold', fontstyle='italic')
plt.xlabel('X-axis', fontsize=16, color='darkgreen', fontweight='bold')
plt.ylabel('Y-axis', fontsize=16, color='darkred', fontfamily='serif')

# Show the plot
plt.show()
```

Example 5: Customizing Legend Fonts

```python
# Create a plot with customized legend fonts
plt.plot(x, y, color='c', label='Sine Wave')
plt.title('Plot with Customized Legend Fonts', fontsize=16)
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)
plt.legend(loc='best', fontsize='large', shadow=True, title='Legend Title', title_fontsize='13', frameon=False)

# Show the plot
plt.show()
```

###  Adjusting axis limits and ticks.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

Example 1: Adjusting Axis Limits

```python
# Create a plot with adjusted axis limits
plt.plot(x, y, color='b')
plt.title('Plot with Adjusted Axis Limits')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set x-axis and y-axis limits
plt.xlim(2, 8)  # Set x-axis limits from 2 to 8
plt.ylim(-0.5, 0.5)  # Set y-axis limits from -0.5 to 0.5

# Show the plot
plt.show()
```

Example 2: Customizing Axis Ticks

```python
# Create a plot with customized axis ticks
plt.plot(x, y, color='g')
plt.title('Plot with Customized Axis Ticks')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
# Customize x-axis ticks
plt.xticks(ticks=[0, 2, 4, 6, 8, 10], labels=['0', '2', '4', '6', '8', '10'])

# Customize y-axis ticks
plt.yticks(ticks=np.linspace(-1, 1, 5), labels=['-1', '-0.5', '0', '0.5', '1'])

# Show the plot
plt.show()
```

Example 3: Rotating Axis Tick Labels

```python
# Create a plot with rotated axis tick labels
plt.plot(x, y, color='r')
plt.title('Plot with Rotated Axis Tick Labels')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set x-axis and y-axis limits
plt.xlim(0, 10)
plt.ylim(-1, 1)

# Rotate x-axis tick labels by 45 degrees
plt.xticks(rotation=45)

# Rotate y-axis tick labels by 90 degrees
plt.yticks(rotation=90)

# Show the plot
plt.show()
```

 Example 4: Adding Grid Lines

```python
# Create a plot with grid lines and customized ticks
plt.plot(x, y, color='m')
plt.title('Plot with Grid Lines and Customized Ticks')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set x-axis and y-axis limits
plt.xlim(0, 10)
plt.ylim(-1, 1)

# Add grid lines
plt.grid(True)  # Turn on grid lines

# Customize x-axis and y-axis ticks
plt.xticks(ticks=np.arange(0, 11, 2))  # x-axis ticks every 2 units
plt.yticks(ticks=np.arange(-1, 2, 0.5))  # y-axis ticks every 0.5 units

# Show the plot
plt.show()
```

Example 5: Using Logarithmic Scale for Axes

```python
# Create a plot with logarithmic scale
x_log = np.logspace(0, 2, 100)  # Generate logarithmically spaced data
y_log = np.sin(x_log)

plt.plot(x_log, y_log, color='c')
plt.title('Plot with Logarithmic Scale')
plt.xlabel('X-axis (log scale)')
plt.ylabel('Y-axis')

# Set x-axis to logarithmic scale
plt.xscale('log')

# Customize y-axis ticks
plt.yticks(ticks=np.linspace(-1, 1, 5))

# Show the plot
plt.show()
```
