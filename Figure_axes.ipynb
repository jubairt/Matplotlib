### Creating Figures and Axes in Matplotlib

```python
import matplotlib.pyplot as plt

# Create a figure with a single axes
fig, ax = plt.subplots()

# Plot some data on the axes
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

ax.plot(x, y, marker='o', linestyle='-', color='b')  # Plot data on the axes
ax.set_title('Simple Line Plot')  # Set title for the plot
ax.set_xlabel('X-axis')  # Label for x-axis
ax.set_ylabel('Y-axis')  # Label for y-axis

plt.show()  # Display the figure
```

Figures with Multiple Axes (Subplots)

```python
import matplotlib.pyplot as plt

# Create a figure with two axes (subplots)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))  # 1 row, 2 columns

# Plot data on the first axes (left subplot)
ax1.plot([1, 2, 3], [1, 4, 9], marker='o', color='r')
ax1.set_title('First Plot')

# Plot data on the second axes (right subplot)
ax2.plot([1, 2, 3], [2, 3, 5], marker='x', color='b')
ax2.set_title('Second Plot')

plt.show()
```

Example 1: Line Plot using `plot()`

```python
# Data for the line plot
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a simple line plot
plt.plot(x, y, marker='o', linestyle='-', color='b')

# Add title and labels
plt.title('Line Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

Example 2: Scatter Plot using `scatter()`

```python
# Data for the scatter plot
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a scatter plot
plt.scatter(x, y, color='r')

# Add title and labels
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

Example 3: Bar Plot using `bar()`

```python
# Data for the bar chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]

# Create a bar chart
plt.bar(categories, values, color='g')

# Add title and labels
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show the plot
plt.show()
```

Example 4: Histogram using `hist()`

```python
# Data for the histogram
data = [1, 2, 2, 2, 3, 3, 4, 4, 4, 4, 5, 5, 6, 7, 8, 9]

# Create a histogram
plt.hist(data, bins=5, color='purple', edgecolor='black')

# Add title and labels
plt.title('Histogram Example')
plt.xlabel('Data Bins')
plt.ylabel('Frequency')

# Show the plot

plt.show()
```
