 Creating and customizing heatmaps with imshow() and pcolormesh().

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for imshow
data = np.random.rand(10, 10)  # 10x10 matrix of random values

# Sample data for pcolormesh
x = np.linspace(0, 10, 10)
y = np.linspace(0, 10, 10)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

plt.figure(figsize=(14, 6))
```

1. Creating a Heatmap with imshow()

```python
plt.subplot(1, 2, 1)
plt.imshow(data, cmap='viridis', interpolation='nearest')
plt.title('Heatmap with imshow()')
plt.colorbar(label='Value')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(False)  # No grid for heatmap
```

 2. Customizing Heatmap with imshow()

```python
plt.subplot(1, 2, 2)
plt.imshow(data, cmap='coolwarm', interpolation='nearest', origin='lower')
plt.title('Customized Heatmap with imshow()')
plt.colorbar(label='Value')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(False)  # No grid for heatmap

plt.tight_layout()
plt.show()
```

Creating a Heatmap with pcolormesh()

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.pcolormesh(X, Y, Z, cmap='plasma', shading='auto')
plt.title('Heatmap with pcolormesh()')
plt.colorbar(label='Value')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

4. Customizing Heatmap with pcolormesh()

```python
plt.subplot(1, 2, 2)
plt.pcolormesh(X, Y, Z, cmap='cividis', shading='auto', edgecolor='black')
plt.title('Customized Heatmap with pcolormesh()')
plt.colorbar(label='Value')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.tight_layout()
plt.show()
```

### Introduction to 3D plotting with Axes3D.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Sample data for 3D plotting
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
```

1. Creating a 3D Surface Plot

```python
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(121, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title('3D Surface Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
```

2. Creating a 3D Scatter Plot

```python
x = np.random.standard_normal(100)
y = np.random.standard_normal(100)
z = np.random.standard_normal(100)

fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(122, projection='3d')
ax.scatter(x, y, z, c='r', marker='o')

ax.set_title('3D Scatter Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Adjust layout
plt.tight_layout()
plt.show()
```
